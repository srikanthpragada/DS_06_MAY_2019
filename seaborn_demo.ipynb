```python
import seaborn as sb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sb.set(style="whitegrid")
```

```python
tips = sb.load_dataset("tips")
```

```python
tips.head()
```

```python
sb.relplot(x='total_bill',y ='tip', hue="time", col="sex", data=tips)
```

```python
sb.relplot(x='total_bill',y ='tip', hue="time", col="sex", data=tips  , kind="line")
```

```python
sb.boxplot(x="day",y="total_bill",data=tips, hue="sex", meanline=True, showmeans=True)
```

```python
sb.barplot(x="day",y="total_bill",hue="sex",data=tips, ci=False)
plt.gcf().set_size_inches(15,5)
```

```python
sb.barplot(x="sex",y="total_bill",hue="day",data=tips, ci=False)
plt.gcf().set_size_inches(10,5)
```

```python
sb.catplot(x="day",y="total_bill",hue="sex", col="time", data=tips, ci=False, kind='bar')
plt.gcf().set_size_inches(10,5)
```

```python
sb.countplot(x="day", hue='smoker', data=tips)
```

```python
sb.pairplot(data=tips, vars=['total_bill','tip'])
```

```python
sb.regplot(x="total_bill",y='tip', data=tips)
```

```python
sb.lmplot(x="total_bill",y='tip', col="sex", hue="smoker", data=tips)
```

```python
sb.residplot(x="total_bill",y='tip', data=tips)
```

```python
data = tips.pivot_table (index = "day", columns = "time", values="total_bill",aggfunc=np.sum)
sb.heatmap(data)
```
