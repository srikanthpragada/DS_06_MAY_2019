```python
def hasUSD(lst):
    for d in lst:
        if d['code'] == 'USD':
            return True
    else:
        return False 
```

```python
hasUSD([ {'code':'INR'}, {'code':'USD'}])
```

```python
countries.loc[countries['currencies'].apply(hasUSD),['name','currencies']]
```

```python
def get_currency_codes(lst):
    codes = []
    for d in lst:
        if d['code'] != None:
            codes.append(d['code'])
        
    # print(codes)    
    if len(codes) == 0:
        return ""
    else:
        return ",".join(codes)        
```

```python
countries['curcodes'] = countries['currencies'].apply(get_currency_codes)
```

```python
import matplotlib.pyplot as plt
```

```python
tc = countries.sort_values (by='population', ascending=False).head(10)[['population','area','name']]
plt.scatter(x = tc ['area'], y = tc ['population'])
plt.gcf().set_size_inches (15,8) # gcf () get current figure and change its size
plt.xlabel("Area")
plt.ylabel("Population")
for i in tc.index:
   plt.annotate(tc.loc[i,'name'],xy=(tc.loc[i,'area'], tc.loc[i,'population']))
```
